```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import cv2
import torch.nn as nn
import torch.nn.functional as F
import torch
import torchvision 
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
def create_training_data():  # this will save the image array and its category into a new array
    training_data = []

    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass

    return np.asarray(training_data, dtype="object")
DATADIR = 'D:/flowerData/flowers'
CATEGORIES = ["daisy", "dandelion", "rose", "sunflower", "tulip"]
IMG_SIZE = 224
training_data = create_training_data()  # a total of 4317 rows, each row's 1st column is the actual image, in the form of 2D array, the 2nd column is the category number
```

```python
DATADIR = 'D:/flowerData/flowers'
CATEGORIES = ["daisy", "dandelion", "rose", "sunflower", "tulip"]
IMG_SIZE = 224
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
transform = transforms.Compose([transforms.RandomHorizontalFlip(),
                                transforms.RandomRotation(0.2),
                                transforms.ToTensor(),
                                transforms.Resize((IMG_SIZE,IMG_SIZE))
                               ])
dataset = torchvision.datasets.ImageFolder(root = DATADIR,
                                           transform = transform)
```

```python
train, val = torch.utils.data.random_split(dataset, [3000, len(dataset)-3000])  # spilt the data to training set as well as test set
```

```python
train_loader = torch.utils.data.DataLoader(dataset = train,
                                           batch_size = 32, 
                                           shuffle = True)
val_loader = torch.utils.data.DataLoader(dataset = val,
                                         batch_size = 32, 
                                         shuffle = True) 
```

```python
# model = nn.Sequential(  # this is the detail of AlexNet
#     nn.Conv2d(3,96,kernel_size=11,stride=4,padding=2),nn.ReLU(),  # 96 is the output channels because in AlexNet paper he write 96 
#     nn.MaxPool2d(kernel_size=3,stride=2),
#     nn.Conv2d(96,128*2,kernel_size=5,padding=2),nn.ReLU(),
#     nn.MaxPool2d(kernel_size=3,stride=2),
#     nn.Conv2d(128*2,192*2,kernel_size=3,padding=1),nn.ReLU(),
#     nn.Conv2d(192*2,192*2,kernel_size=3,padding=1),nn.ReLU(),
#     nn.Conv2d(192*2,128*2,kernel_size=3,padding=1),nn.ReLU(),
#     nn.MaxPool2d(kernel_size=3,stride=2), 
#     nn.Flatten(),
#     nn.Linear(6*6*256,2048*2),nn.ReLU(),nn.Dropout(p=0.5),
#     nn.Linear(2048*2,2048*2),nn.ReLU(),nn.Dropout(p=0.5),
#     nn.Linear(2048*2,len(CATEGORIES)),nn.ReLU(),
# )
class ownAlex(nn.Module):
    def __init__(self):
        super(ownAlex, self).__init__()
        self.features = nn.Sequential(         # this is the detail of AlexNet
            nn.Conv2d(3,96,kernel_size=11,stride=4,padding=2),nn.ReLU(),  # 96 is the output channels because in AlexNet paper he write 96 
            nn.MaxPool2d(kernel_size=3,stride=2),
            nn.Conv2d(96,128*2,kernel_size=5,padding=2),nn.ReLU(),
            nn.MaxPool2d(kernel_size=3,stride=2),
            nn.Conv2d(128*2,192*2,kernel_size=3,padding=1),nn.ReLU(),
            nn.Conv2d(192*2,192*2,kernel_size=3,padding=1),nn.ReLU(),
            nn.Conv2d(192*2,128*2,kernel_size=3,padding=1),nn.ReLU(),
            nn.MaxPool2d(kernel_size=3,stride=2), 
        )
        #self.avgpool = nn.AdaptiveAvgPool2d((6, 6))
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(6*6*256,2048*2),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(2048*2,2048*2),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.5),
            nn.Linear(2048*2,5),
            #nn.ReLU(inplace=True)
        )

    def forward(self, x: torch.Tensor):
        x = self.features(x)
        x = self.classifier(x)
        return x
    
model = ownAlex()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
#optimizer = torch.optim.SGD(model.parameters(), lr=0.03)
model
```

```python
total_step = len(train_loader)
Loss = []
Acc = []
Val_Loss = []
Val_Acc = []

for epoch in range(3):
    acc = 0
    val_acc = 0
    for i, (images, labels) in enumerate(train_loader):  # sweep through the entire training set
        model.train()
        images = images.to(device)
        labels = labels.to(device)
        #optimizer.zero_grad()  # empty the previous gradient
        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)
#         # Backward and optimize, # another word for optimize is also called move the next step/going to next best position
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        # Checking accuracy
        preds = outputs.data.max(dim=1,keepdim=True)[1]
        acc += preds.eq(labels.data.view_as(preds)).cpu().sum()
    acc = acc/len(train_loader.dataset) * 100
    # After go through entire training set, now we can use the knowledge we learning, to practice with the test set
    for i, (images, labels) in enumerate(val_loader):
        model.eval()
        images = images.to(device)
        labels = labels.to(device)
        # Forward pass
        outputs = model(images)
        val_loss = criterion(outputs, labels)
        # Checking accuracy
        preds = outputs.data.max(dim=1,keepdim=True)[1]
        val_acc += preds.eq(labels.data.view_as(preds)).cpu().sum()
    val_acc = val_acc/len(val_loader.dataset) * 100
    print("Epoch {} =>  loss : {loss:.2f};   Accuracy : {acc:.2f}%;   Val_loss : {val_loss:.2f};   Val_Accuracy : {val_acc:.2f}%".format(epoch+1, loss=loss.item(), acc=acc, val_loss=val_loss.item(), val_acc=val_acc))
    Loss.append(loss)
    Acc.append(acc)
    Val_Loss.append(val_loss)
    Val_Acc.append(val_acc)
```

```python
alexnet = torchvision.models.alexnet(pretrained=False)
alexnet.classifier[6].out_features = 5
for param in alexnet.features.parameters(): 
    param.requires_grad = True
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(alexnet.parameters(), lr=0.001)
```

```python
# using sample AlexNet
total_step = len(train_loader)
Loss = []
Acc = []
Val_Loss = []
Val_Acc = []

for epoch in range(3):
    acc = 0
    val_acc = 0
    for i, (images, labels) in enumerate(train_loader):
        alexnet.train()
        images = images.to(device)
        labels = labels.to(device)
        # Forward pass
        outputs = alexnet(images)
        loss = criterion(outputs, labels)
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        # Checking accuracy
        preds = outputs.data.max(dim=1,keepdim=True)[1]
        acc += preds.eq(labels.data.view_as(preds)).cpu().sum()
    acc = acc/len(train_loader.dataset) * 100
    
    for i, (images, labels) in enumerate(val_loader):
        alexnet.eval()
        images = images.to(device)
        labels = labels.to(device)
        # Forward pass
        outputs = alexnet(images)
        val_loss = criterion(outputs, labels)
        # Checking accuracy
        preds = outputs.data.max(dim=1,keepdim=True)[1]
        val_acc += preds.eq(labels.data.view_as(preds)).cpu().sum()
    val_acc = val_acc/len(val_loader.dataset) * 100
    print("Epoch {} =>  loss : {loss:.2f};   Accuracy : {acc:.2f}%;   Val_loss : {val_loss:.2f};   Val_Accuracy : {val_acc:.2f}%".format(epoch+1, loss=loss.item(), acc=acc, val_loss=val_loss.item(), val_acc=val_acc))
    Loss.append(loss)
    Acc.append(acc)
    Val_Loss.append(val_loss)
    Val_Acc.append(val_acc)
```
